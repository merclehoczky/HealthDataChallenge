# Filter  parameters

```python
conditions = ['StrGlo', 'WD', 'WVv']
filtered = w19_hr.loc[~w19_hr['Parameter'].isin(conditions)]
```

```python
filtered.head(50)
```

# Create new dataframe with locations sorted under timepoints

```python
w19_new = pd.pivot(filtered, index = ['Datum', 'Standort'], columns = ['Parameter'], values = ['Wert'])
```

```python
w19_new
```

```python
w19_new.columns = w19_new.columns.droplevel()
```

# Averaged values under each timepoint

Save dates

```python
dates = w19_new.reset_index(inplace=False)
dates = dates['Datum']
dates = dates.drop_duplicates()
dates = dates.to_frame()
dates = dates.reset_index(drop = True)
dates.head(50)
```

Average values

```python
w19_new.groupby(np.arange(len(w19_new))//3).mean()
```

Join dataframes

```python
w19_new = w19_new.reset_index(drop=True)
dates = dates.reset_index(drop=True)
```

```python
dates.head(50)
```

```python
w19 = pd.merge(w19_new, dates, left_index = True, right_index = True)
w19 = w19[['Datum', 'Hr', 'RainDur', 'T', 'WVs', 'p']]
```
