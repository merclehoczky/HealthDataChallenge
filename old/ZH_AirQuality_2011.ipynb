Filtering for PM10 (fine dust indicator)

```python
aq_hr_PM10 = aq_hr[aq_hr.Parameter.str.startswith('PM10')]
aq_hr_PM10
```

```python
aq_hr_PM10.describe()
```

```python
aq_hr_PM10.hist(bins = 200)
```

Typically, fine dust concentration over 100 µg/m3 causes visible pollution

```python
aq_hr_PM10[aq_hr_PM10.Wert > 100 ].Wert.hist(bins = 100)
```

```python
np.count_nonzero(aq_hr_PM10.Wert>100)
```
